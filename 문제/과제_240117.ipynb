```python
# 1
import time

def time_decorator(func):
    def wrapper():
        start_time = time.time()
        result = func()
        end_time = time.time()
        print(f"함수가 {end_time - start_time} 초가 걸렸습니다.")
        return result
    return wrapper

@time_decorator
def sum_to_100():
    res = 0
    for i in range(1, 101):
        res += i
    return res

print(sum_to_100())
# 2
def gen():
    i = 0
    while i != 100:
        i += 2
        yield i

for i in gen():
    print(i)
# 3
def divide(x, y):
    try:
        return x / y
    except ZeroDivisionError:
        print('0으로 나눌 수 없습니다.')

print(divide(4, 3))
print(divide(2, 0))
```
